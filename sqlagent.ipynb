```python
# import libraries
import os

from langchain_groq import ChatGroq
from langchain_community.utilities import SQLDatabase
from langchain_community.agent_toolkits import create_sql_agent
```

```python
# set up LLM 
os.environ["GROQ_API_KEY"] = "gsk_
llm = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0
)
```

```python
# test LLM
response = llm.invoke("Say hello in one sentence.")
print(response.content)
```

```python
# connect langchain to mysql database
from urllib.parse import quote_plus

password = quote_plus("#Pineapple123")

db = SQLDatabase.from_uri(
    f"mysql+pymysql://root:{password}@localhost:3306/student_db"
)
```

```python
# testing connection
print(db.get_usable_table_names())
```

```python
# creating sql agent 
agent = create_sql_agent(
    llm=llm,
    db=db,
    agent_type="tool-calling",
    verbose=True
)
```

```python
# testing sql agent
agent.invoke("How many students are in the students table?")
```

```python
agent.invoke("What is the average final grade G3 in the students table?")
```

```python
agent.invoke(
    "What is the average final grade G3 for students with internet access compared to students without internet access?"
)
```
